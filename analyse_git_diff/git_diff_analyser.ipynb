```python
import pandas as pd
import csv
pd.options.display.max_colwidth = None
```

# get the data from txt file
__ is column separator, data lines are joined in gvim (diff determines beginning of one record)
my_cols gives max possible col size and ensures that all rows with all cols will be imported

```python
my_cols = ['files', 'diff_reason', 'info1', 'info2', 'info3', 'info4']
df = pd.read_csv("git_diff_joined_lines_utf8.txt", sep = '__', names = my_cols, engine = 'python')
```

# remove data where similarity index 100% for column diff
they're just renamed and we're not interested in them

```python
len(df[df.diff_reason == 'similarity index 100%'])
```

```python
df = df[df.diff_reason != 'similarity index 100%']

len(df)
```

# find unique reasons

```python
# Split on space and take the first element:
df['diff_reason'].str.split().str[0].unique()
```

## how many of each unique reasons we have

```python
# create new column with only short reasons
df['short_reason'] = df['diff_reason'].str.split().str[0]

df.groupby('short_reason')['files'].nunique()
```

reasons for checking:

index = files differ, names the same

deleted = file in oglasi deleted, exists only in Arhiva

new = file in oglasi, but not in Arhiva

similarity = non 100%, manual check of files?

### investigate what are the reasons exactly, prepare keywords from info columns
merge lists given by [-3:] and such into strings
check if there's any other reason except those expected

```python
# last word, expected 'differ', when short == index
df['imgdiff'] = df['info1'].str.split().str[-1].str.join('')
df[(df.short_reason == 'index') & (df.imgdiff != 'differ')]
```

```python
# last 3 words, expected 'and /dev/null differ', when short == deleted
df['in_arhiva'] = df['info2'].str.split().str[-3:].str.join(' ')
df[(df.short_reason == 'deleted') & (df.in_arhiva != 'and /dev/null differ')]
```

```python
# first 4 words, expected 'Binary files /dev/null and', when short == new
df['in_oglasi'] = df['info2'].str.split().str[:4].str.join(' ')

df[(df.short_reason == 'new') & (df.in_oglasi != 'Binary files /dev/null and')]
```

```python
df['files_quote'] = df['files'].str.extract('a/(.*)" "b/')
df['files_noquote'] = df['files'].str.extract('a/(.*)b/')
```

```python
df['files2'] = df['files_quote']
```

```python
df.loc[df['files_quote'].isnull(), 'files2'] = df['files_noquote']
```

```python
with pd.option_context('max_rows', None):
  display(df[df.short_reason == 'new'][:10])
```

```python
df[df.short_reason == 'similarity']['files'].to_csv(r"out_similar_provjeri_slike.txt", 
                                                    index = False, header = False, 
                                                    quoting = csv.QUOTE_NONE, escapechar = "\\")
```

```python
df[df.short_reason == 'new']['files2'].to_csv(r"out_new-samo_u_oglasu.txt", 
                                              index = False, header = False, 
                                              quoting = csv.QUOTE_NONE, escapechar = "\\")
```

```python
df[df.short_reason == 'deleted']['files2'].to_csv(r"out_del-samo_u_arhivi_prebaci_u_oglas.txt", 
                                                  index = False, header = False, 
                                                  quoting = csv.QUOTE_NONE, escapechar = "\\")
```

```python
df[df.short_reason == 'index']['files2'].to_csv(r"out_index_analiziraj.txt", 
                                                index = False, header = False,
                                                quoting = csv.QUOTE_NONE,  escapechar = "\\")
```
